```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
progress_df = pd.read_csv("results/it_net2/progress.csv")

train_loss = list(progress_df["train_loss"])[1:]
val_loss = list(progress_df["val_loss"])[1:]
steps = list(progress_df["step"])[1:]
```

```python
plt.plot(steps, train_loss, label="train_loss")
plt.plot(steps, val_loss, label="val_loss")
plt.xlabel("steps")
plt.ylabel("loss")
plt.legend()
plt.show()
```
